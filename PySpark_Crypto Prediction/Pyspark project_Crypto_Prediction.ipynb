### **Creating categorizations and grouping columns**


We have created 4 categorizations based on the various types of columns in the data frame. These categorizations serve as the building blocks for our pipeline and enable us to call on different parts of the dataframe with ease.

#### Identifier column

```python
# Creating the identifier column

IDENTIFIER_COLUMN = ["ID"]
print(IDENTIFIER_COLUMN)
```

#### Binary column

```python
# Identifying binary type columns from the data frame, df_crypto, and converting them into categorical data type
# by defining a function and using a for loop to parse through the dataframe

from pyspark.sql.functions import col
from pyspark.ml.feature import StringIndexer

# Defining a function to identify binary columns (i.e. those columns with only 2 values (0 or 1))
def is_binary_column(df_crypto, column_name):
    distinct_values = df_crypto.select(column_name).distinct().count()
    return distinct_values == 2

# Initializing an empty list to store binary column names
BINARY_COLUMN = []

# Iterating through all columns of the DataFrame
for column in df_crypto.columns:
    if is_binary_column(df_crypto, column):

        # Appending the binary column name to the list
        BINARY_COLUMN.append(column)

        # Checking if the column is already 'string' type, if not, casting it as such
        if df_crypto.schema[column].dataType != "string":
            # If not, casting it to string type
            df_crypto = df_crypto.withColumn(column, col(column).cast("string"))

        # Converting identified column to categorical type
        output_column_name = column + "_index"  # Modify output column name to ensure unique names are given to each column
        indexer = StringIndexer(inputCol=column, outputCol=output_column_name, handleInvalid="keep").fit(df_crypto)
        df_crypto = indexer.transform(df_crypto)

        # Droping the original binary column to avoid redundancy df_crypto
        df_crypto = df_crypto.drop(column)

# Show the updated DataFrame
df_crypto.show()
```

```python
# Print the list of binary column names
print("Number of Binary Columns (including target variable):", len(BINARY_COLUMN))
print("They are:", BINARY_COLUMN)
```

```python
# Although the for loop required the reformatted columns have unique names,
# we have standardized the names in the dataframe crypto to retain the original column names

columns_to_rename = df_crypto.columns

# Filtering columns that end with "_index"
columns_to_rename = [col for col in columns_to_rename if col.endswith("_index")]

# Iterating over the columns to rename and remove "_index"
for col in columns_to_rename:
    old_col_name = col
    new_col_name = col[:-len("_index")]  # Remove "_index" from the column name
    df_crypto = df_crypto.withColumnRenamed(old_col_name, new_col_name)
```

## **Building Pipeline**

```python
#import pipeline
from pyspark.ml import Pipeline

#import  pyspark.ml.feature which contains Imputer,VectorAssembler,MinMaxScaler
import pyspark.ml.feature as MF
```

```python
#Stage 1
# Creating a loop to add suffix to continuous columns
continuous_imputed = []
for i in range(len(CONTINUOUS_COLUMN)):
  continuous_imputed.append(CONTINUOUS_COLUMN[i]+'_i')
continuous_imputed
```

```python
#imputer_continuous replaces nulls in continuous columns with mean values
imputer_continuous = MF.Imputer(
strategy="mean",
inputCols=CONTINUOUS_COLUMN,
outputCols=continuous_imputed,
)
```

```python
# Stage 2
# Creating a loop to add suffix to binary columns
binary_imputed = []
for i in range(len(BINARY_COLUMN)):
  binary_imputed.append(BINARY_COLUMN[i]+'_i')
binary_imputed
```

```python
#imputer_binary replaces nulls in binary columns with mode values
imputer_binary = MF.Imputer(
strategy="mode",
inputCols=BINARY_COLUMN,
outputCols=binary_imputed,
)
```

```python
# Stage 3
#vector assembler used to create a dense vector
continuous_assembler = MF.VectorAssembler(
inputCols=continuous_imputed,
outputCol="continuous_vec",
)
```

```python
#Stage 4
#minmax standardization implemented on dense vector
continuous_scaler = MF.MinMaxScaler(
inputCol="continuous_vec",
outputCol="continuous_scaled",
)
```

```python
# Stage 5
#New Vector Assembler used to create a dense vector combining binary and continuous columns
preml_assembler = MF.VectorAssembler(
                   inputCols=binary_imputed + ["continuous_scaled"],
                   outputCol="features")
```

```python
#  Bring all stages
my_pipeline = Pipeline(
stages=[imputer_continuous, imputer_binary, continuous_assembler, continuous_scaler, preml_assembler]
)
```

# 5. Concluding Observations




**Observation 1**

We explored many ways to treat this dirty dataset - particularly the null values. As there were many null values present when we eyeballed the data, we initially tried to replace all the null values with zero for the purpose of modelling. However, there was some information to be gained from considering imputation as a method. For this reason, we decided to impute null values based on the type of column we were dealing with. For binary column types, we imputed based on the mode of the column. For continuous column types, we imputed based on the mean of the column.

**Observation 2**

Given the size of the dataset being relatively small, we took a collective decision to retain as much information as possible. To do this, we avoided dropping any rows or columns and tried to build base models with all information available. We made better progress when we ran a stacking model on top of the predictions from our base models.

**Observation 3**

Among the models we explored, we found that hypertuning and playing with the parameters proved critical to tweaking the F1 score and improving the information gleaned out of each model. Some models worked really well in explaining this dataset such as Random Forest Classifiers and Decision Trees. Interstinglyy, other models such as MLP, Gradient Boosted Trees or SVC models, did not perform as well. The worst performing model we had was the SVC model.

**Observation 4**

Another observation we made was that a single model by itself would not yield a weighted F-1 score higher than 0.719. This was likely due to the fact that the information captured from the complex dataset was limited. Our success lay in the fact that we combined results from multiple models using a stacking ensemble which boosted the accuracy of the target class prediction. The only drawback to our approach is the loss of interpretability that is associated with ensemble models.

Finally, we also explored strengthening our model by using cross validation and hyperparameter tuning. However, the computational time was quite high.
